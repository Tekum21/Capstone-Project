```python
# sorting individuals affected in descending order
Report=Report.sort_values(by="Individuals Affected",ascending=False)
#Top 20 entities affected
Report.head(20)
```

```python
# Finding top 10 states with most breaches. 
D = Report.groupby('State',as_index=None).agg({"Individuals Affected":"sum"})
D.sort_values("Individuals Affected", ascending= False,ignore_index=True).head(10)
```

```python
# Ranking the type of data breach by the number of individuals affected. 
D = Report.groupby('Type of Breach',as_index=None).agg({"Individuals Affected":"sum"})
D.sort_values("Individuals Affected", ascending= False,ignore_index=True).head(10)
```

```python
# Creating a new sub dataset that contains the total number of affected individuals for each month.
# This is to permit us do a closer analysis by months.
New_data = pd.read_excel(r"C:\Users\tnjok\OneDrive\Desktop\Datasets\Breach Summary.xlsx")
New_data
```

```python
sum(New_data["Totoal Individuals Affected"])
```

```python
# create the time series plot
fig,ax=plt.subplots(figsize=(10,8))
sns.lineplot(x = "Month", y = "Totoal Individuals Affected",data = New_data)
plt.xticks(rotation = 90)
plt.show()
```

```python
# create the time series plot
fig,ax=plt.subplots(figsize=(10,8))
sns.lineplot(x = "Breach Month & Year", y = "Totoal Individuals Affected",data = New_data)
plt.title("Individuals Affected",color='red',fontsize=14)
plt.xticks(rotation = 25)

plt.show()
```

```python
#Investigating the distribution of breaches.
fig,ax=plt.subplots(figsize=(10,8))
Rev=New_data["Totoal Individuals Affected"].values
ax=sns.distplot(Rev,color="blue")
#Calculating the mean
mean=(New_data["Totoal Individuals Affected"].mean())
print(mean)
#ploting the mean
plt.axvline(mean,color="red",label="Av Individuals Affected")
plt.title('Histogram showing the distribution of the number of breaches',color='red',fontsize=14)
plt.legend()
plt.grid()
```

```python
#Example1: Bar plot showing the Type of Breaches and the Affected Individuals
fig,ax=plt.subplots(figsize=(10,8))
ax=sns.barplot(y="Totoal Individuals Affected",x="Month",data=New_data,ci=None,
               palette = "Reds")
plt.title("Total Number of Breaches Reported in Their Respective Months and Years", color='blue',fontsize=14) 
plt.xticks(rotation = 60)
plt.show(fig)
```
